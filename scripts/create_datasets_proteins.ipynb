```python
import os
from collections import Counter
import pandas as pd
from Bio import SeqIO
os.chdir('../data')
```

```python
def count_aminoacid(protein):
    """Подсчёт аминокислот в белке, 
    argument: protein - The string sequence of amino acids in a protein"""
    amino_counter = Counter()
    for aminoacid in protein:
        amino_counter[aminoacid] += 1
    return amino_counter

def organism_name(index):
    return ListOrganism.iloc(0)[index][1]
def organism_type(index):
    return ListOrganism.iloc(0)[index][0]
def organism_stein(index):
    return ListOrganism.iloc(0)[index][2]
def organism_filename(index):
    return ListOrganism.iloc(0)[index][3]

def add_proteom(Type, Organism , Strain, file_name):
    for record in SeqIO.parse(file_name + '.fasta', "fasta"):
        # Добавление в список подсчёт аминокислот у протеома организма из файла
        CounterAminoacid = count_aminoacid(record.seq)
        ListCountAminoacid = []
        for aminoacid in 'ABCDEFGHIJKLMNOPQRSTUVWYZX':
            ListCountAminoacid.append(CounterAminoacid[aminoacid])
        ListProteins.loc[len(ListProteins)] = [Organism, record.name, Strain, Type] + ListCountAminoacid
```

```python
ListOrganism = pd.read_csv('ListOrganism.csv')
ListOrganism
```

```python
ListProteins = pd.DataFrame({'organism' : [], 'strain' : [], 'type' : [], 'protein' : []})
for aminoacid in 'ABCDEFGHIJKLMNOPQRSTUVWYZX': ListProteins['freq_' + aminoacid] = []
```

```python
for index in range(len(ListOrganism)):
    print(index)
    add_proteom(organism_type(index), organism_name(index), organism_stein(index), organism_filename(index))
ListProteins
```

```python
ListProteins.to_csv('ListProteins.csv')
```
