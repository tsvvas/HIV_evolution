```python
import sys
sys.path = ["../utils"] + sys.path
import pandas
import os
from Bio import SeqIO
import Bio
from Bio.Data import CodonTable
import matplotlib.pyplot as plt
import seaborn as sns
from aa_properties import *
from translate_dna import *
```

```python
def calc_feature(method="hydro", type_haplo="major", file_name="haplotypes_p1_V3.fasta"):
    max_value_freq, min_value_freq = {}, {}
    hydro, unique_days, size_freq = [], [], []
    dict_function = {'hydro' : score_hydrophobicity_sequence, 'kidera' : score_sequence, 
                     'prop' : aaprop_sequence}
    function_score = dict_function[method]
    
    if type_haplo == "major":
        
        for record in SeqIO.parse("haplotypes_p1_V3.fasta", "fasta"):
            current_freq = int(record.id.split('_')[3][:-1])
            current_day = int(record.id.split('_')[1])
            max_value_freq[current_day] = max(current_freq,
                                                          max_value_freq.setdefault(current_day, 0))
            
        for record in SeqIO.parse("haplotypes_p1_V3.fasta", "fasta"):
            current_freq = int(record.id.split('_')[3][:-1])
            current_day = int(record.id.split('_')[1])
            if max_value_freq[current_day] == current_freq:
                hydro.append(function_score(translate_dna(record.seq)))
                unique_days.append(current_day)
                size_freq.append(current_freq)
    
    elif type_haplo == "minor":
        
        for record in SeqIO.parse("haplotypes_p1_V3.fasta", "fasta"):
            current_freq = int(int(record.id.split('_')[3][:-1]))
            current_day = int(record.id.split('_')[1])
            min_value_freq[current_day] = min(current_freq,
                                                          min_value_freq.setdefault(current_day, 100))
            
        for record in SeqIO.parse("haplotypes_p1_V3.fasta", "fasta"):
            current_freq = int(int(record.id.split('_')[3][:-1]))
            current_day = int(record.id.split('_')[1])
            if min_value_freq[current_day] == current_freq:
                hydro.append(function_score(translate_dna(record.seq)))
                unique_days.append(current_day)
                size_freq.append(int(current_freq))
    
    elif type_haplo == "all":
        
        for record in SeqIO.parse("haplotypes_p1_V3.fasta", "fasta"):
            current_freq = int(int(record.id.split('_')[3][:-1]))
            current_day = int(record.id.split('_')[1])
            hydro.append(function_score(translate_dna(record.seq)))
            unique_days.append(current_day)
            size_freq.append(current_freq)

    return (unique_days, hydro, size_freq)
```

```python
def draw_feature(x, y, size=None, method='all'):
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set_xlabel('Days after infection')
    ax.set_ylabel('Hydrophobycity') 
    ax.set_xlim([0, max(max(x) + 100, 300)])
    ax.set_ylim([min(y) - 5, max(y) + 5])
    if method == 'all':
        sns.boxplot(x=x, y=y, palette='Blues_d')
    else:
        sns.scatterplot(x=x, y=y, size=size, palette='Blues_d', sizes=(100, 200))
```

```python
x, y, size = calc_feature(type_haplo='all', method='hydro')
draw_feature(x, y)
```

```python
x, y, size = calc_feature(type_haplo='major', method='hydro')
draw_feature(x, y, size, method='any')
```

```python
x, y, size = calc_feature(type_haplo='minor', method='hydro')
draw_feature(x, y, size, method='any')
```
